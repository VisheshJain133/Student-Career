```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import evaluate
import torch
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/student_responses_updated_dataset.csv")

# Check the columns in the dataset
print(df.columns)

# Use 'What would you like to be?' as the label column
df['label'] = df['What would you like to be?']  # Replace with the correct column for labels

# Convert the label column to integers
df['label'] = df['label'].astype('category').cat.codes  # Converts text labels to categorical integer values

# Combine 'Explanation' and 'Hobbies' columns into one text input
df['input_text'] = df['Explanation'] + " " + df['Hobbies']

# Tokenization using BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Function to tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['input_text'], padding='max_length', truncation=True)

# Convert pandas DataFrame to Hugging Face Dataset
dataset = Dataset.from_pandas(df[['input_text', 'label']])

# Tokenize the dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split dataset into train and test sets using Hugging Face's built-in function
train_test_split_dataset = tokenized_dataset.train_test_split(test_size=0.2)

# Extract train and test datasets
train_dataset = train_test_split_dataset['train']
eval_dataset = train_test_split_dataset['test']

# Load the BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(df['label'].unique()))  # Adjust num_labels to the number of unique labels

# Define training arguments with progress bar disabled
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',  # Directory for storing logs
    logging_steps=10,  # Log every 10 steps
    disable_tqdm=True  # Disable progress bar to avoid the error
)

# Use the evaluate library for metrics
accuracy_metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy_metric.compute(predictions=predictions, references=labels)

# Create a custom trainer to collect metrics
class CustomTrainer(Trainer):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.train_losses = []
        self.eval_losses = []
        self.eval_accuracies = []

    def log(self, logs: dict):
        super().log(logs)
        if "loss" in logs:
            self.train_losses.append(logs["loss"])
        if "eval_loss" in logs:
            self.eval_losses.append(logs["eval_loss"])
        if "eval_accuracy" in logs:
            self.eval_accuracies.append(logs["eval_accuracy"])

# Define the custom trainer
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    compute_metrics=compute_metrics,  # Adding metrics for evaluation
)

# Train the model
trainer.train()

# Evaluate the model
results = trainer.evaluate()

# Print the evaluation metrics
print(results)

# Plotting the results
def plot_metrics(trainer):
    epochs = list(range(1, len(trainer.eval_losses) + 1))

    # Create a figure and axis
    fig, ax1 = plt.subplots()

    # Plot loss
    color = 'tab:red'
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss', color=color)
    ax1.plot(epochs, trainer.eval_losses, color=color, label='Eval Loss')
    ax1.tick_params(axis='y', labelcolor=color)

    # Create a twin axis for accuracy
    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Accuracy', color=color)
    ax2.plot(epochs, trainer.eval_accuracies, color=color, label='Eval Accuracy')
    ax2.tick_params(axis='y', labelcolor=color)

    # Add titles and display the plot
    plt.title("Loss and Accuracy over Epochs")
    fig.tight_layout()  # To prevent clipping of y-labels
    plt.show()

# Plot the results
plot_metrics(trainer)
```

```python
# Import necessary libraries
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Define the career_labels mapping
career_labels = {
    "Engineer": 0,
    "Doctor": 1,
    "Artist": 2,
    "Scientist": 3,
    "Teacher": 4,  # Ensure you have all labels defined
}

# Load pre-trained model and tokenizer (modify model name as needed)
model_name = "bert-base-uncased"  # Replace with your model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(career_labels))

# Define the device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Make predictions on new student responses
def predict_new_text(text):
    # Put the model in evaluation mode
    model.eval()

    # Tokenize the input text and move to the appropriate device
    encoding = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    encoding = {key: val.to(device) for key, val in encoding.items()}  # Move input tensors to the same device

    # Move the model to the same device
    model.to(device)

    with torch.no_grad():  # Disable gradient calculation for inference
        outputs = model(**encoding)

    logits = outputs.logits
    prediction = torch.argmax(logits, dim=1).item()  # Get the predicted class index

    # Debugging: Print the prediction and logits
    print(f"Logits: {logits}")  # Logits from the model
    print(f"Predicted class index: {prediction}")  # Index of the predicted class

    # Map the predicted integer back to the career label
    for label, index in career_labels.items():
        if index == prediction:
            return label

    return "Prediction out of range"  # In case the prediction doesn't match any label

# Example usage
#new_text = "I enjoy solving problems and want to help people improve their lives saving their life."

new_text = "I am fascinated by biology and want to help people by finding cures for diseases and improving healthcare."
#new_text = "I love solving complex technical problems and creating innovative solutions that make life easier for people."
#new_text = "I am passionate about justice and want to defend the rights of those who cannot defend themselves."
#new_text = "I enjoy sharing knowledge and helping others learn new things, and I want to inspire future generations."
#new_text = "I have a deep curiosity about the universe and want to explore new frontiers in scientific research."

predicted_label = predict_new_text(new_text)
print(f"The predicted career label is: {predicted_label}")
```

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
import numpy as np

# Get predictions on the validation dataset
predictions = trainer.predict(eval_dataset) # Use eval_dataset which was used for training

# Print the predictions object to check its contents
print("Predictions object:", predictions)

# Extract predicted and true labels
preds = predictions.predictions
labels = predictions.label_ids

# Print shapes of predictions and labels
print("Predictions shape:", preds.shape if preds is not None else "None")
print("Labels shape:", labels.shape if labels is not None else "None")

# Check if the labels and preds are not None
if labels is None or preds is None:
    raise ValueError("No labels or predictions found.")

# Calculate predicted class labels
predicted_classes = np.argmax(preds, axis=-1)

# Calculate accuracy
accuracy = accuracy_score(labels, predicted_classes)

# Calculate precision, recall, F1-score for each class
precision, recall, f1, _ = precision_recall_fscore_support(labels, predicted_classes, average='macro')

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
